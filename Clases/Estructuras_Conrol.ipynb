```python
def lista_1(l1):
  #Queremos que nos de el promedio
  v=0
  for i in range(0,len(l1)): #Para i en el rango, de 0 a un elemento de la lista
    v+=l1[i]
    i+=1
  return v/len(l1)

lista_1([0,1,2,3])
```

```python
def prod_punto(l1,l2): #Definimos el producto punto
  #Queremos que nos de el PRODUCTO PUNTO
 m,n=len(l1),len(l2) #Definimos 2 vectores
 if m!=n: #Si el vector m es distinto (!=) del vector n, devuelve...
     return 'NOOO'
 else: #Sino...
  s=0 #Empezamos de un resultado igual a 0
  for i in range(n): #Para i en el rango de la cantidad de elemtos del vector n
    s+=l1[i]*l2[i] #Tenemos que el resultado "s" es igual al producto de las entradas iesimas de cada vector (x_1*y_1 + x_2*y_2+...x_i*y_i)
  return s #Sino, devuelve "s"
print(prod_punto([1,2],[1,2])) #Definimos los elementos de cada vector
```

```python
def f(x):
  return x**2 #Definimos una funcion f, que depende de x
def g(x):
  return (x+2) #Definimos la funcion g, que depende de x
def fog(x):
  return f(g(x)) #Definimos una funcion compuesta y la llamamos fog(x), pues dependera de la variable x
print(fog(1)) #Corremos la funcion compuesta para x=1
```

```python
#Ejemplo de Yield
def numeros_naturales(): #Definimos a la funcion de los numeros naturales
  n=1 #Establecemos como 1 el primer elemento
  while True: #Creamos un ciclo para n, usando la funcion "Yield"
    yield n #La funcion Yield, es para hacer suesiones al vuelo
    n+=1 #a partir de n=1, sumamos de 1 en 1 un elemento
print(numeros_naturales())
```

```python
for natural in numeros_naturales(): #Elegimos elementos de la funcion previamente definida
  print(natural)
  if natural>9: #Si la cantidad de elementos naturales excede a 9 elementos (partiendo de 0), nos detenemos con "break"
    break
```

**Fibonacci:**

```python
def memoisacion(f):
    memo = {}
    def auxiliar(x):
        if x not in memo:
            memo[x] = f(x)
        return memo[x]
    return auxiliar

def fibo_recursivo(n):
    if n <= 1:
        return n
    return g(n - 1) + g(n - 2)

g = memoisacion(fibo_recursivo)

for i in range(10):
    print(i, g(i))
```
