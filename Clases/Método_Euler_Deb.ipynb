*Articulo Deb*

Para reducir la varianza Deb, reducimos el error N, para ello se hace uso de funciones de densidad probabilistica (PDF), tal que, si queremos integrar una funcion, debemos escoger una PDF parecida a nuestra funcion, tal que haya una relacion entre ellas, esta es:
$$\frac{f(x)}{g(x)} = c $$

Como queremos que la PDF sea parecida a nuestra funcion que queremos integrar, debemos buscar que $c ≈ 1 $. Con $g(x)$ nuestra PDF.

La integral queda de la forma: $$ \int_{a}^{b} f(x) dx = \int_{a}^{b} \left(\frac{f(x)}{g(x)}\right) g(x) dx$$

**Rangos infinitos**

Por resolver $$ \frac{dx}{dt} = \frac{1}{x²+t²} $$ con el cambio de variable
 $$ g(x,u) = (1-u)^{-2} \frac{1}{x^2+u^2/(1-u)^2} $$

**Ecuaciones diferenciales ordinarias**

*Método de Euler:*



```python
from pylab import *
```

```python
#Trabajamos con un cambio de variable
def g(x,u):
  return 1/(x*+2 * (1-u)**2 + u**2)

def Euler_r_inf(g,x0,a,b,N):
  h=(b-a)/N
  x=x0
  lista_u=arange(a,b+h,h)
  lista_x,lista_t= [], []
  for u in lista_u:
    lista_t.append(u/(1-u))
    lista_x.append(x)
    x+=h*g(x,u) #Euler
  return lista_t,lista_x
t,x=Euler_r_inf(g,1,0,1,100)
plot(t,x)
```

```python

######################## Euler ####
def Euler_1d(f,x0,a,b,N):
   h = (b-a)/N
   x=x0
   lista_t=arange(a,b+h,h)
   lista_x= []
   for i in lista_t:
      lista_x.append(x)

      x+=h*f(x,i) #EULER
   return lista_t,lista_x
#######################

f=lambda x,t:-x**3 + sin(t)
t,x=Euler_1d(f,0,0,10,100)
plot(t,x)
```
